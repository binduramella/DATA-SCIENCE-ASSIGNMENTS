```python
#importing library
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# **Dataset:**

```python
dff = pd.read_excel('Online retail.xlsx', names=['items']) # Assuming file is in the same directory

# Assuming the separator is ';' and the column to split is 'Description'
df_split = dff['items'].str.split(',', expand=True)

df_split.head()
```

```python
dataset = df_split.replace(np.nan, '', regex=True)
dataset
```

```python
dataset
transactions = []
for i in range(0, 7500):
    transactions.append([str(dataset.values[i,u]) for u in range(0, 19)])
```

```python
groceries_series  = pd.DataFrame(pd.Series(transactions))
groceries_series
groceries_series.columns = ["business"]
groceries_series
```

```python
X = groceries_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

# **Data Preprocessing:**

```python

df=X.copy()
```

```python

#information
df.info()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#dropping duplicate
df1=df.copy()
df1=df1.drop_duplicates()
df1
```

```python
#creating dummies
df1=pd.get_dummies(df1)
df1= df1.astype(int)
df1
```

# **Apriori Algorithm**

```python
# Applying Apriori Algorithm
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(df1, min_support=0.01,max_len=4,use_colnames = True)
frequent_itemsets
```

```python
#finding according to lift
# Calculate the number of itemsets in frequent_itemsets
num_itemsets = len(frequent_itemsets)

# Pass num_itemsets to association_rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1, num_itemsets=num_itemsets)

rules.head(20)
rules.sort_values('lift', ascending=False).head(10)
```

```python
# High chances of if and then
rules.sort_values('lift',ascending = False)[0:20]
```

```python

#Highest chances of lift of rare items
rules[rules.lift>3]
```

```python
#these are likely to buy together
# Calculate the number of itemsets in frequent_itemsets
num_itemsets = len(frequent_itemsets)

# Pass num_itemsets to association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5, num_itemsets=num_itemsets)

rules.head(20)
rules.sort_values('confidence', ascending=False).head(10)
```

# **INTERVIEW QUESTIONS**

1.What is Lift and Why is it Important?
Lift measures how much more likely the consequent occurs given the antecedent, compared to random chance (Lift=Confidence/Support of Consequent).

It identifies meaningful rules by filtering out those based solely on high support items.

A lift > 1 indicates a positive association, making it crucial for finding significant relationships.

2.What are Support and Confidence, and How Are They Calculated?
Support: The fraction of transactions containing an itemset (Support=Transactions with Itemset/Total Transactions).

Confidence: The likelihood of the consequent given the antecedent (Confidence= Support of Antecedent/Support of Rule).

Both measure the frequency and reliability of association rules in the dataset.

3.What Are Some Limitations of Association Rule Mining?
It can generate too many rules, making it hard to identify meaningful patterns.

Rare but significant associations might be missed due to strict support thresholds.

High computational cost for large datasets or itemsets with many combinations.
