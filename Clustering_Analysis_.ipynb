## **Data Preprocessing:**

```python
# importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data set
dataset="EastWestAirlines.xlsx"
df = pd.read_excel(dataset, sheet_name='data', engine='openpyxl')
df
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df_cleaned = df.drop(columns=['ID#'])
df_cleaned
```

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

# Remove outliers from the dataset
df_cleaned = remove_outliers_iqr(df_cleaned)
df_cleaned
```

```python
# Normalize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cleaned)
df_scaled = pd.DataFrame(df_scaled, columns=df_cleaned.columns)
```

## **Hierarchical Clustering:**

Different linkage criteria (Ward, Complete, Average, Single) were tested.

## **DBSCAN Clustering:**

Different values of epsilon (eps) and min_samples were tested.

Some parameter combinations resulted in too many noise points (-1 cluster), meaning DBSCAN struggles to define meaningful clusters for this dataset.

The Silhouette Scores for DBSCAN show that performance varies greatly depending on parameter selection.

Unlike K-Means, DBSCAN does not assume spherical clusters, making it more robust for detecting outliers.
