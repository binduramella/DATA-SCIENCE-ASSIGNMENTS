```python
import pandas as pd
import matplotlib.pyplot as plt

numerical_cols = [4, 8, 9, 10, 11, 12]  # Indices of numerical columns
data_frame = df.iloc[:, numerical_cols]

# Creating individual boxplots
for col in data_frame.columns:
    plt.figure(figsize=(6, 4))  # Adjusting figure size
    plt.boxplot(data_frame[col], vert=False)  # Horizontal boxplot for better label visibility
    plt.title(f'Boxplot of {col}')
    plt.xlabel('Value')
    plt.show()
```
